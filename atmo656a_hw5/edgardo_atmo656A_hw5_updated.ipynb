Computing T and R by using Meador & Weaver (1979) - Quadrature Method with $\tau$ = 4, $\omega_{0}$ = 1, $g$ = 0.8 and $\mu_{0}$ = 1:

```python
tau_p = 4
mu0   = 1 # 45 degrees for sza
w0    = 0.9999
g     = 0.8
```

```python
# Quadrature Method. Coefficients from Table 1 of Meador & Weaber (1979)
gamma1 = ((3**(1/2))/2)*(2 - w0*(1 + g))
gamma2 = ((3**(1/2))*(w0/2))*(1 - g)
gamma3 = (1/2)*(1 - (3**(1/2))*g*mu0)
gamma4 = 1 - gamma3

alpha1 = gamma1*gamma4 + gamma2*gamma3
alpha2 = gamma1*gamma3 + gamma2*gamma4
k      = (gamma1**2 - gamma2**2)**(1/2)

# ===============
# Reflectivity from Eq. 14 of Meador & Weaber (1979):
R_a = w0/((1 - (k**2)*(mu0**2))*(((k + gamma1)*np.exp(k*tau_p)) + ((k - gamma1)*np.exp(-k*tau_p))))
R_b = (1 - k*mu0)*(alpha2 + k*gamma3)*np.exp(k*tau_p)
R_c = (1 + k*mu0)*(alpha2 - k*gamma3)*np.exp(-k*tau_p)
R_d = 2*k*(gamma3 - alpha2*mu0)*np.exp(-tau_p/mu0)

R = R_a*(R_b - R_c - R_d)

# Transmissivity from Eq. 14 of Meador & Weaber (1979):
T_a = R_a
T_b = (1 + k*mu0)*(alpha1 + k*gamma4)*np.exp(k*tau_p)
T_c = (1 - k*mu0)*(alpha1 - k*gamma4)*np.exp(-k*tau_p)
T_d = 2*k*(gamma4 + alpha1*mu0)*np.exp(tau_p/mu0)

T = np.exp(-tau_p/mu0)*(1 - T_a*(T_b - T_c - T_d))

print('Transmissivity',round(T,3))
print('Reflectivity',round(R,3))
```

#### Estimating Transmissivity and Reflectivity for stratus cloud layer

We need to find the total water column in molecules $cm^{-1}$. We can start estimating the saturation vapor pressure $e_{sat}$ at a temperature $T = 280 K$ by using Clausius-Clapeyron equation (see eq. 4.38 of Murry L. Salby "Physics of the Atmosphere and Climate" as reference):

$$e_{sat} = e_{0}\;exp\left[\frac{L_v}{R_v}\left(\frac{1}{T_0}-\frac{1}{T}\right)\right]$$

where $e_{0}$ and $T_{0}$ can be a reference saturation pressure and temperature point respectively (6.11 mbar and 273 K for instance), $L_v$ is the latent heat of vaporization ($2.5\times10^{6}\;Jkg^{-1}$) and $R_v$ the gas constant for 1 kg of water vapor ($462\;JK^{-1}kg^{-1}$)

```python
e0 = 611    # Pa
T0 = 273    # K
Lv = 2.5e6  # J kg-1
Rv = 462    # J K-1 kg-1
T  = 280    # K   
e_sat = e0*np.exp((Lv/Rv)*((1/T0)-(1/T)))
print(round(e_sat,1),'Pa')
```

Then, we can compute the total water column $u$ in units of molecules $cm^{-1}$, for a depth of $\Delta z = 250\;m$ and temperature $T = 280\;K$ by using the ideal gas law:

$$u = \frac{e_{sat}}{RT}N_{A}\Delta z$$

where $R$ is the gas constant (8.314 $JK^{-1}mol^{-1}$) and $N_A$ is the Avogadro's Number ($6.022\times 10^{23}$ molecules $mol^{-1}$).
