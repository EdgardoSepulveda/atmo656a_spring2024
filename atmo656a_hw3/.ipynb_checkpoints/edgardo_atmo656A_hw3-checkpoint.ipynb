Computing T and R by using Meador & Webaer (1979) and Liou (1973):

```python
tau = np.arange(0, 10.1, 0.1)
mu0 = np.array([0.9, 0.5, 0.1])
w0  = np.array([0.9999, 0.9])

g   = 0.75
F0  = 1360 # value doesn't matter, will cancelled out eventually (specifically for R & T calculation; Not for Flux calc.)

R = np.zeros((len(tau), len(mu0), len(w0), 3)) # Optical Depth x mu0 x wo x 3 methdos (2 M&W methods and Liou)
T = np.zeros((len(tau), len(mu0), len(w0), 3)) # Optical Depth x mu0 x wo x 3 methdos (2 M&W methods and Liou)
for w0_i in range(len(w0)):
    for mu0_i in range(len(mu0)):
        for tau_i in range(len(tau)):
            for method in range(2):
                if method == 0: # Eddington Method. Coefficients from Table 1 of Meador & Weaber (1979)
                    gamma1 = (1/4)*(7 - w0[w0_i]*(4 + 3*g))
                    gamma2 = (-1/4)*(1 - w0[w0_i] * (4 - 3*g))
                    gamma3 = (1/4)*(2 - 3*g*mu0[mu0_i])
                    gamma4 = 1 - gamma3
                else: # Quadrature Method. Coefficients from Table 1 of Meador & Weaber (1979)
                    gamma1 = ((3**(1/2))/2)*(2 - w0[w0_i]*(1 + g))
                    gamma2 = ((3**(1/2))*(w0[w0_i]/2))*(1 - g)
                    gamma3 = (1/2)*(1 - (3**(1/2))*g*mu0[mu0_i])
                    gamma4 = 1 - gamma3
    
                alpha1 = gamma1*gamma4 + gamma2*gamma3
                alpha2 = gamma1*gamma3 + gamma2*gamma4
                k      = (gamma1**2 - gamma2**2)**(1/2)
            
                # ===============
                # Re[flectivity from Eq. 14 of Meador & Weaber (1979):
                R_a = w0[w0_i]/((1 - (k**2)*(mu0[mu0_i]**2))*(((k + gamma1)*np.exp(k*tau[tau_i])) + ((k - gamma1)*np.exp(-k*tau[tau_i]))))
                R_b = (1 - k*mu0[mu0_i])*(alpha2 + k*gamma3)*np.exp(k*tau[tau_i])
                R_c = (1 + k*mu0[mu0_i])*(alpha2 - k*gamma3)*np.exp(-k*tau[tau_i])
                R_d = 2*k*(gamma3 - alpha2*mu0[mu0_i])*np.exp(-tau[tau_i]/mu0[mu0_i])

                R[tau_i, mu0_i, w0_i, method] = R_a*(R_b - R_c - R_d)
               
                # Transmissivity from Eq. 14 of Meador & Weaber (1979):
                T_a = R_a
                T_b = (1 + k*mu0[mu0_i])*(alpha1 + k*gamma4)*np.exp(k*tau[tau_i])
                T_c = (1 - k*mu0[mu0_i])*(alpha1 - k*gamma4)*np.exp(-k*tau[tau_i])
                T_d = 2*k*(gamma4 + alpha1*mu0[mu0_i])*np.exp(tau[tau_i]/mu0[mu0_i])

                T[tau_i, mu0_i, w0_i, method] = np.exp(-tau[tau_i]/mu0[mu0_i])*(1 - T_a*(T_b - T_c - T_d))
                
            # Liou Method (See Liou (1974))

            mu1 = np.sqrt(1/3) # Between eq. 6 and 7 of Liou (1974)
            w1  = 3*g*w0[w0_i] # From where?
            k1  = (1/mu1)*((1 - w0[w0_i])*(1 - w1*mu1**2))**(1/2) # Eq. 11 from Liou (1974)
            # W1(mu1) Eq. 12 from Liou (1974) (Positive mu0 for upward):
            W1_pos = (1/(1 + mu1*k1))*(w0[w0_i] - w1*(1 - w0[w0_i])*mu1/k1)
            # W1(-mu1) Eq. 12 from Liou (1974) (Negative mu0 for downward):
            W1_neg = (1/(1 + (-mu1)*k1))*(w0[w0_i] - w1*(1 - w0[w0_i])*(-mu1)/k1)
            # Z(mu1) Eq. 13 from Liou (1974) (Positive mu0 for upward):
            Z_pos  = (mu0[mu0_i]*F0/4)*(((mu1 - mu0[mu0_i])*(w0[w0_i] - w1*(1 - w0[w0_i])*mu1*mu0[mu0_i]))/(mu1**2*(1 - k1**2*mu0[mu0_i]**2)))
            # Z(-mu1) Eq. 13 from Liou (1974) (Negative mu0 for downward):
            Z_neg  = (mu0[mu0_i]*F0/4)*((((-mu1) - mu0[mu0_i])*(w0[w0_i] - w1*(1 - w0[w0_i])*(-mu1)*mu0[mu0_i]))/((-mu1)**2*(1 - k1**2*mu0[mu0_i]**2)))
            # C+ Eq. 10 from Liou (1974):
            C_pos  = -(Z_neg + Z_pos*np.exp(-tau[tau_i]/mu0[mu0_i]))
            # C+ Eq. 10 from Liou (1974):
            C_neg  = -(Z_neg - Z_pos*np.exp(-tau[tau_i]/mu0[mu0_i]))
            # A+ Eq. 9 from Liou (1974):
            A_pos  = W1_neg + W1_pos*np.exp(-k1*tau[tau_i])
            # A- Eq. 9 from Liou (1974):
            A_neg  = W1_neg - W1_pos*np.exp(-k1*tau[tau_i])                
            # L1 Eq. 7 from Liou (1974):
            L_pos  = (1/2)*((C_pos/A_pos) + (C_neg/A_neg))
            # L-1 Eq. 8 from Liou (1974):
            L_neg  = (1/2)*((C_pos/A_pos) - (C_neg/A_neg))*np.exp(-k1*tau[tau_i])      
            
            # ===============
            # Upward Flux (Eq. 5 from Liou (1974)):
            F_up = 2*np.pi*mu1*(L_pos*W1_pos + L_neg*W1_neg + Z_pos) # what about the exponential terms? Why applying the exponential shift the curve to zero?
            #F_up = 2*np.pi*mu1*(L_pos*W1_pos*np.exp(-k1*tau[tau_i]) + L_neg*W1_neg*np.exp(k1*tau[tau_i]) + Z_pos*np.exp(-tau[tau_i]/mu0[mu0_i]))
            
            # Downward Flux (Eq. 6 from Liou (1974)):
            F_dn = (-2*np.pi*mu1)*(L_pos*W1_neg*np.exp(-k1*tau[tau_i]) + L_neg*W1_pos*np.exp(k1*tau[tau_i]) + Z_neg*np.exp(-tau[tau_i]/mu0[mu0_i])) - np.pi*mu0[mu0_i]*F0*np.exp(-tau[tau_i]/mu0[mu0_i])
            
            #Reflectivity
            R[tau_i, mu0_i, w0_i, 2] = F_up/(np.pi*mu0[mu0_i]*F0)
            
            #Transmissivity
            T[tau_i, mu0_i, w0_i, 2] = abs(F_dn)/(np.pi*mu0[mu0_i]*F0)
```
