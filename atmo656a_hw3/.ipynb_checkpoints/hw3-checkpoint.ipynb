```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import AxesGrid
from math import gamma
from itertools import islice
from netCDF4 import Dataset
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
import glob
import string
import copy
from matplotlib.colors import LogNorm
import time
```

```python
#Medor and Weaver
#define array
tau=range(1,11);

mu0=np.array([0.1,0.5,0.9]); w0=np.array([0.9,0.99999])
#R_a=np.zeros((10,3,2)); R_b=np.zeros((10,3,2)); R_c=np.zeros((10,3,2)); R_d=np.zeros((10,3,2))
R=np.zeros((10,3,2,2))
R_liou=np.zeros((10,3,2))
#T_a=np.zeros((10,3,2)); T_b=np.zeros((10,3,2)); T_c=np.zeros((10,3,2)); T_d=np.zeros((10,3,2))
T=np.zeros((10,3,2,2))
T_liou=np.zeros((10,3,2))


#define parameter
g=0.75
F0=1360.  #value doesn't matter, will cancelled out eventually (specifically for R & T calculation; Not for Flux calc.)

#loop thru. the
for nn in range(2):                #method controller
    for ii in range(2):            #w0
        for jj in range(3):        #mu0
            for kk in range(10):   #tau
            
                #--switch between methods
                if nn == 0:    #Eddington
                    gamma1=(1./4.)*(7-w0[ii]*(4+3*g))
                    gamma2=(-1./4.)*(1-w0[ii]*(4-3*g))
                    gamma3=(1./4.)*(2-3*g*mu0[jj])
                    gamma4=1-gamma3
                    
                if nn == 1:  #Quadrature
                    gamma1=(3**(1./2.)/2)*(2-w0[ii]*(1+g))
                    gamma2=(3**(1./2.)*w0[ii]/2)*(1-g)
                    gamma3=(1./2.)*(1-3**(1./2.)*g*mu0[jj])
                    gamma4=1-gamma3
    
                alpha1=gamma1*gamma4+gamma2*gamma3
                alpha2=gamma1*gamma3+gamma2*gamma4
                kappa=(gamma1**2-gamma2**2)**(1./2.)
            
                #write equation
                #===============
                #Re[flectivity
                R_a=w0[ii]/((1-kappa**2*mu0[jj]**2)*(((kappa+gamma1)*np.exp(kappa*tau[kk]))+\
                                                     ((kappa-gamma1)*np.exp(-kappa*tau[kk]))))
                R_b=(1-kappa*mu0[jj])*(alpha2+kappa*gamma3)*np.exp(kappa*tau[kk])
                R_c=(1+kappa*mu0[jj])*(alpha2-kappa*gamma3)*np.exp(-kappa*tau[kk])
                R_d=2*kappa*(gamma3-alpha2*mu0[jj])*np.exp(-tau[kk]/mu0[jj])
                ##
                R[kk,jj,ii,nn]=R_a*(R_b-R_c-R_d)
                #===============
                #Transmissivity
                T_a=R_a
                T_b=(1+kappa*mu0[jj])*(alpha1+kappa*gamma4)*np.exp(kappa*tau[kk])
                T_c=(1-kappa*mu0[jj])*(alpha1-kappa*gamma4)*np.exp(-kappa*tau[kk])
                T_d=2*kappa*(gamma4+alpha1*mu0[jj])*np.exp(tau[kk]/mu0[jj])
                ##
                T[kk,jj,ii,nn]=np.exp(-tau[kk]/mu0[jj])*(1-T_a*(T_b-T_c-T_d))
                #===============
                
                #Liou

                #define parameter
                mu1=np.sqrt(1/3)
                w1=3*g*w0[ii]
                kappa1=(1/mu1)*((1-w0[ii])*(1-w1*mu1**2))**(1/2)
                #W1(mu1)
                W1_pos=(1/(1+mu1*kappa1))*(w0[ii]-w1*(1-w0[ii])*mu1/kappa1)
                #W1(-mu1)
                W1_neg=(1/(1+(-mu1)*kappa1))*(w0[ii]-w1*(1-w0[ii])*(-mu1)/kappa1)
                #Z(mu1)
                Z_pos=(mu0[jj]*F0/4)*\
                (((mu1-mu0[jj])*(w0[ii]-w1*(1-w0[ii])*mu1*mu0[jj]))/(mu1**2*(1-kappa1**2*mu0[jj]**2)))
                #Z(-mu1)
                Z_neg=(mu0[jj]*F0/4)*\
                ((((-mu1)-mu0[jj])*(w0[ii]-w1*(1-w0[ii])*(-mu1)*mu0[jj]))/((-mu1)**2*(1-kappa1**2*mu0[jj]**2)))
                #C-plus
                C_pos=-(Z_neg+Z_pos*np.exp(-tau[kk]/mu0[jj]))
                #C-minus
                C_neg=-(Z_neg-Z_pos*np.exp(-tau[kk]/mu0[jj]))
                #A-plus
                A_pos=W1_neg+W1_pos*np.exp(-kappa1*tau[kk])
                #A-minus
                A_neg=W1_neg-W1_pos*np.exp(-kappa1*tau[kk])                
                #L-plus1
                L_pos=(1/2)*((C_pos/A_pos)+(C_neg/A_neg))
                #L-minus1
                L_neg=(1/2)*((C_pos/A_pos)-(C_neg/A_neg))*np.exp(-kappa1*tau[kk])              
                #=====write equations
                #Upward Flux
                F_up=2*np.pi*mu1*\
                (L_pos*W1_pos+L_neg*W1_neg+Z_pos)
                #Downward Flux
                F_dn=(-2*np.pi*mu1)*\
                (L_pos*W1_neg*np.exp(-kappa1*tau[kk])+L_neg*W1_pos*np.exp(kappa1*tau[kk])+Z_neg*np.exp(-tau[kk]/mu0[jj]))\
                -np.pi*mu0[jj]*F0*np.exp(-tau[kk]/mu0[jj])
                #Reflectivity
                R_liou[kk,jj,ii]=F_up/(np.pi*mu0[jj]*F0)
                #Transmissivity
                T_liou[kk,jj,ii]=abs(F_dn)/(np.pi*mu0[jj]*F0)
                #------
                #print("Fup=",F_up,"  Fdn=",F_dn)
                #print("R=",R_liou,"  T=",T_liou," R+T=",R_liou+T_liou)
```

```python
#Plotting
fsize=15

fig=plt.figure(figsize=(20,15))
#fig, axs = plt.subplots(3,2,sharex=True,gridspec_kw={'hspace': 0.2, 'wspace': 0.2})
ax1=fig.add_subplot(3,2,1)
ax1.set_title('R & T: '+'  $\omega_0$='+str(w0[0])+';  $\mu_0$='+str(mu0[0]),fontsize=fsize)
ax1.plot(tau,T[:,0,0,0],'-bo',label='T - Eddi')
ax1.plot(tau,T[:,0,0,1],'--bd',label='T - Quad')
ax1.plot(tau,R[:,0,0,0],'-ro',label='R - Eddi')
ax1.plot(tau,R[:,0,0,1],'--rd',label='R - Quad')
ax1.legend(loc='center right', shadow=False, fontsize=10)
ax1.set_xticks(np.arange(min(tau),max(tau)+1,1.0))

ax2=fig.add_subplot(3,2,3)
ax2.set_title('R & T: '+'  $\omega_0$='+str(w0[0])+';  $\mu_0$='+str(mu0[1]),fontsize=fsize)
ax2.plot(tau,T[:,1,0,0],'-bo',label='T - Eddi')
ax2.plot(tau,T[:,1,0,1],'--bd',label='T - Quad')
ax2.plot(tau,R[:,1,0,0],'-ro',label='R - Eddi')
ax2.plot(tau,R[:,1,0,1],'--rd',label='R - Quad')
ax2.set_xticks(np.arange(min(tau),max(tau)+1,1.0))

ax3=fig.add_subplot(3,2,5)
ax3.set_title('R & T: '+'  $\omega_0$='+str(w0[0])+';  $\mu_0$='+str(mu0[2]),fontsize=fsize)
ax3.plot(tau,T[:,2,0,0],'-bo',label='T - Eddi')
ax3.plot(tau,T[:,2,0,1],'--bd',label='T - Quad')
ax3.plot(tau,R[:,2,0,0],'-ro',label='R - Eddi')
ax3.plot(tau,R[:,2,0,1],'--rd',label='R - Quad')
ax3.set_xticks(np.arange(min(tau),max(tau)+1,1.0))

ax11=fig.add_subplot(3,2,2)
ax11.set_title('R & T: '+'  $\omega_0$='+str(w0[1])+';  $\mu_0$='+str(mu0[0]),fontsize=fsize)
ax11.plot(tau,T[:,0,1,0],'-bo',label='T - Eddi')
ax11.plot(tau,T[:,0,1,1],'--bd',label='T - Quad')
ax11.plot(tau,R[:,0,1,0],'-ro',label='R - Eddi')
ax11.plot(tau,R[:,0,1,1],'--rd',label='R - Quad')
ax11.legend(loc='center right', shadow=False, fontsize=10)
ax11.set_xticks(np.arange(min(tau),max(tau)+1,1.0))

ax22=fig.add_subplot(3,2,4)
ax22.set_title('R & T: '+'  $\omega_0$='+str(w0[1])+';  $\mu_0$='+str(mu0[1]),fontsize=fsize)
ax22.plot(tau,T[:,1,1,0],'-bo',label='T - Eddi')
ax22.plot(tau,T[:,1,1,1],'--bd',label='T - Quad')
ax22.plot(tau,R[:,1,1,0],'-ro',label='R - Eddi')
ax22.plot(tau,R[:,1,1,1],'--rd',label='R - Quad')
ax22.set_xticks(np.arange(min(tau),max(tau)+1,1.0))

ax33=fig.add_subplot(3,2,6)
ax33.set_title('R & T: '+'  $\omega_0$='+str(w0[1])+';  $\mu_0$='+str(mu0[2]),fontsize=fsize)
ax33.plot(tau,T[:,2,1,0],'-bo',label='T - Eddi')
ax33.plot(tau,T[:,2,1,1],'--bd',label='T - Quad')
ax33.plot(tau,R[:,2,1,0],'-ro',label='R - Eddi')
ax33.plot(tau,R[:,2,1,1],'--rd',label='R - Quad')
ax33.set_xticks(np.arange(min(tau),max(tau)+1,1.0))
plt.show()
fig.savefig('HW3_1_XjZheng.png')
```
