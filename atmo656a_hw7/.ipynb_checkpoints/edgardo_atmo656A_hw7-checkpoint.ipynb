PyMieScatt package is used to compute mie scattering

Importing libraries:

```python
import numpy as np
import pandas as pd
import time
#from scipy.special import jv, yv
from scipy.integrate import trapz
#from scipy import stats          #needed for lognormal distribution
#import warnings
import PyMieScatt as pms
import matplotlib.pyplot as plt
#import matplotlib.ticker as ticker
```

We need to compute the total extinction coefficient $\overline{Q}_{ext}$ of a distribution of sizes, by integrating the different extinction coefficient $Q_{ext}$ from each particle in the sample, weighted by the lognormal size distribution $n(r)$ (The probability that we can find that particular size in the sample):

$$\overline{Q}_{ext} = \frac{1}{N_0}\int_{0}^{\infty} Q_{ext}(m,\lambda,r)n(r)dr$$

where $Q_{ext}$ is the extinction coefficient for a single particle of size $r$, refractive index $m$, and interacting with a radiation beam of wavelength $\lambda$. $N_0$ is the total number of aprticles in the sample.

```python
m       = 1.3873 + 0.0066j          # Refractive Index of particle
wl      = 500                       # wavelength in nm (unit requested by PyMieScatt)
gsd     = [1.1, 1.4, 1.8, 2.1, 2.5] # Geometric Standard Deviation of size distribution
gmean   = np.arange(10,10010,10)    # Geometric mean of size distribution in nm (unit requested by PyMieScatt)
npart   = 10000                     # Number of particles
nMedium = 1                         # Refractive Index of medium
```

Let's compute the extinction coefficient of a single particle first, by using the PyMieScatt.MieQ() function (similar to HW6), and check if this is working:

```python
Q_ext = []
for ri in gmean:
    result = pms.MieQ(m, wl, ri, nMedium)
    Q_ext.append(result[0])
```

Size parameter:

```python
x = 2*np.pi*gmean/wl
```

```python
plt.plot(x, Q_ext)
plt.xlabel('Size parameter', fontsize = 16)
plt.ylabel(r'$Q_{ext}$', fontsize = 16)
```

Now, let's check the different lognormal size distributions, for each geometric standard deviation and some mean sizes selected:

$$ n(r)dr = \frac{1}{\sqrt{2\pi}\;ln(\sigma_g)} exp\left(-\frac{(ln(r)-ln(r_g))^2}{2(ln(\sigma_g))^2}\right) $$

```python
fig, ax = plt.subplots(5, 1, figsize = (10, 10), sharex = True)

r_space = np.logspace(-1, 6, 100) # 100 different sizes.
r_means = np.array([10, 100, 1000, 2000, 5000, 10000]) # Mean sizes for each lognormal distribution to plot 0.01 to 10 um):
for gsd_i in range(len(gsd)):
    for r_i in range(len(r_means)):
        lognorm = (1/(np.sqrt(2*np.pi)*np.log(gsd[gsd_i]))) * np.exp( -((np.log(r_space)-np.log(r_means[r_i]))**2)/(2*np.log(gsd[gsd_i])**2))
        ax[gsd_i].plot(r_space, lognorm)
    ax[gsd_i].set_xscale('log')
    if gsd_i == len(gsd)-1: ax[gsd_i].set_xlabel('r (nm)', fontsize = 16)
    ax[gsd_i].set_ylabel('n(r)', fontsize = 16)
    ax[gsd_i].text(0.95, 0.95, r'$\sigma_g$ = '+str(gsd[gsd_i]), 
                   ha = 'right', va = 'top', transform = ax[gsd_i].transAxes, fontsize = 16)
```

We can see that smaller geometric standard deviations need higher resolution of sizes, but a smaller range. We can, for example, select a resolution of 60 sizes for the $\sigma_g = 2.5$ and range from 0.6 nm to 200,000 nm, but a resolution of 200 sizes for the smallest $\sigma_g$ and a range from $10^{0.8}$ nm to $10^{4.2}$ nm. The total calculation of extinction coefficients is reduced from 100*5 to (200+70+60+50+40) = 480 different size points. This is not enough.

We can also reduce the calculation range per mean size, from the whole size range, to a range of mean  2*std (only inside the "bell")

```python
def find_nearest(array, value):
    idx = (np.abs(array - value)).argmin()
    return idx
```

Now it takes like 8 seconds:

```python
start = time.time()
Q_ext_sd = np.zeros((len(gmean),len(gsd)))
for gsd_i in range(len(gsd)):
    ranges_i = [0.8, 0.6, 0.1, 0, -0.2]
    ranges_f = [4.2, 4.5, 4.8, 5, 5.3]
    resolution = [200, 70, 60, 50, 40]
    r_space  = np.logspace(ranges_i[gsd_i], ranges_f[gsd_i], resolution[gsd_i])
    start = time.time()
    for r_i in range(len(gmean)):
        lognorm = (1/(np.sqrt(2*np.pi)*np.log(gsd[gsd_i]))) * np.exp( -((np.log(r_space)-np.log(gmean[r_i]))**2)/(2*np.log(gsd[gsd_i])**2))
        new_initial = find_nearest(r_space, 10**(np.log10(gmean[r_i]) - 2*np.log10(gsd[gsd_i])))
        new_final   = find_nearest(r_space, 10**(np.log10(gmean[r_i]) + 2*np.log10(gsd[gsd_i])))
        sub_range   = r_space[new_initial:new_final]
        sub_log     = lognorm[new_initial:new_final]
        nrqri   = []
        for logri in range(len(sub_range)):
            mie_result = pms.MieQ(m, wl, sub_range[logri], nMedium)
            nrqri.append(sub_log[logri]*mie_result[0])
        Q_ext_sd[r_i, gsd_i] = trapz(nrqri)
end = time.time()
print(end - start,'seconds')
```

```python
magic_factor = 1#0.16 # Why?? I think this is the factor related with N0
Q_ext_sd_scaled = Q_ext_sd*magic_factor

plt.plot(x, Q_ext, label = 'single particle', linewidth = 1.5)
plt.plot(x, Q_ext_sd_scaled[:,0], label = r'$\sigma_g$ = 1.1', linewidth = 1.5)
plt.plot(x, Q_ext_sd_scaled[:,1], label = r'$\sigma_g$ = 1.4', linewidth = 1.5)
plt.plot(x, Q_ext_sd_scaled[:,2], label = r'$\sigma_g$ = 1.8', linewidth = 1.5)
plt.plot(x, Q_ext_sd_scaled[:,3], label = r'$\sigma_g$ = 2.1', linewidth = 1.5)
plt.plot(x, Q_ext_sd_scaled[:,4], label = r'$\sigma_g$ = 2.5', linewidth = 1.5)

plt.ylabel(r'$Q_{ext}$', fontsize = 12)
plt.xlabel('Size parameter', fontsize = 12)

#plt.xscale('log')

#plt.ylim(-0.4, 4.3)

plt.legend(frameon = False, ncol = 1, fontsize = 12, loc='upper left')
```
